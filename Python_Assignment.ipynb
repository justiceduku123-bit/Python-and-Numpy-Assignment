```python
# Quest 1
import math

def phi(n):
    """
    Euler's Totient function: counts numbers <= n that are relatively prime to n
    """
    count = 0
    for k in range(1, n+1):
        if math.gcd(k, n) == 1:
            count += 1
    return count
```

```python
print(phi(9))  
print(phi(10))  
print(phi(1))   
```

```python
# Quest 2
def easter_date(year):
    a = year % 19
    b = year // 100
    c = year % 100
    d = (19 * a + b - (b // 4) - ((b - ((b + 8) // 25) + 1) // 3) + 15) % 30
    e = (32 + 2 * (b % 4) + 2 * (c // 4) - d - (c % 4)) % 7
    m = (a + 11 * d + 22 * e) // 451

    month = (d + e + 114) // 31          
    day = ((d + e + 114) % 31) + 1

    if d == 29 and e == 6:
        month = 4
        day = 19  
    exception_m_values = {2, 5, 10, 13, 16, 21, 24, 39}
    if d == 28 and e == 6 and m in exception_m_values:
        month = 4
        day = 18  

    return month, day


year = 2025
month, day = easter_date(year)
print(f"Easter in {year} falls on: {month}/{day}")
```

```python
# Quest 6
def simplify_text(text):
    text = text.lower()
   
    remove_chars = '.,!?;:"()[]{}<>/@#$%^&*_+=\\|~`'
 
    simplified = ""
    for char in text:
        if char not in remove_chars:
            simplified += char
    
    return simplified
```

```python
# Quest 7
def convert_time(time_str):
   
    hour_str, minute_str = time_str.split(":")
    hour = int(hour_str)
 
    groups = ["A", "B", "C", "D", "E"]
    group_index = hour // 5
    group = groups[group_index]
  
    southern_hour = (hour % 5) + 1
    return f"{southern_hour}{group}:{minute_str}"
```

```python
# Quest 8
def next_name(name):
    letters = list(name)
    i = len(letters) - 1   
    
    while i >= 0:
        if letters[i] != 'Z':
            letters[i] = chr(ord(letters[i]) + 1)
            return ''.join(letters)
        else:
            letters[i] = 'A'
            i -= 1
    return 'A' + ''.join(letters)
```

```python
print(next_name("S"))      # T
print(next_name("CD"))     # CE
print(next_name("AZ"))     # BA
print(next_name("ZZ"))     # AAA
print(next_name("AAB"))    # AAC
```

```python
# Quest 9
def reverse_mirrored(text):
   
    letters = [c for c in text if c.isalpha()]
  
    letters.reverse()
  
    result = []
    letter_index = 0
    
    for c in text:
        if c.isalpha():
            result.append(letters[letter_index])
            letter_index += 1
        else:
            result.append(c) 
    
    return "".join(result)
```

```python
# Quest 11
def closest(numbers, n):
    closest_num = numbers[0]
    
    for num in numbers[1:]:
        if abs(num - n) < abs(closest_num - n):
            closest_num = num
        elif abs(num - n) == abs(closest_num - n) and num < closest_num:
            closest_num = num

    return closest_num
print(closest([5, 10, 14], 12))     
print(closest([1, 8, 20], 10))       
print(closest([-2, -5, 3], 0))       
```

```python
# Quest 14 
def telephone_match(num, s):
    keypad = {
        '2': 'ABC', '3': 'DEF', '4': 'GHI',
        '5': 'JKL', '6': 'MNO', '7': 'PQRS',
        '8': 'TUV', '9': 'WXYZ'
    }

    digits = [c for c in num if c.isdigit()]
  
    s = s.upper()
  
    if len(digits) != len(s):
        return False
   
    for d, letter in zip(digits, s):
        if letter not in keypad.get(d, ''):
            return False
    
    return True
print(telephone_match("468-3437", "INTEGER"))  
print(telephone_match("43556", "HELLO"))        
print(telephone_match("43556", "WORLD"))        
```

```python
 # Quest 18
import random

def triple_shuffle(list1, list2, list3):
    if not (len(list1) == len(list2) == len(list3)):
        raise ValueError("All lists must have the same length")
   
    indices = list(range(len(list1)))
 
    random.shuffle(indices)
  
    list1[:] = [list1[i] for i in indices]
    list2[:] = [list2[i] for i in indices]
    list3[:] = [list3[i] for i in indices]
a = [1, 2, 3, 4, 5]
b = [11, 22, 33, 44, 55]
c = [9, 8, 7, 6, 5]

triple_shuffle(a, b, c)
print(a)  # e.g., [4, 1, 5, 3, 2]
print(b)  # e.g., [44, 11, 55, 33, 22]
print(c)  # e.g., [6, 9, 5, 7, 8]
```

```python
# Quest 22
def find_words(consonants):
    vowels = 'aeiou'
    consonants = consonants.lower()
    matching_words = []

    # Open the word list file
    with open("wordlist.txt", "r") as file:
        for word in file:
            word = word.strip().lower()
    
            word_no_vowels = ''.join([c for c in word if c not in vowels])
           
            if word_no_vowels == consonants:
                matching_words.append(word)
    
    return matching_words

```

```python
# Quest 23
def words_in_word(t):
    t = t.lower()
    words_found = set()  
  
    with open("wordlist.txt", "r") as f:
        word_set = set(word.strip().lower() for word in f)

    for start in range(len(t)):
        for end in range(start + 1, len(t) + 1):
            substring = t[start:end]
            if substring in word_set:
                words_found.add(substring)

```

```python
# Quest 24
def integer_groups(numbers):
    if not numbers:
        return []

    groups = []
    current_group = [numbers[0]]

    for i in range(1, len(numbers)):
        if numbers[i] == numbers[i - 1] + 1:
            current_group.append(numbers[i])
        else:
            groups.append(current_group)
            current_group = [numbers[i]]

    groups.append(current_group)
    return groups
numbers = [1, 2, 5, 6, 7, 10, 11]
print(integer_groups(numbers))
```
