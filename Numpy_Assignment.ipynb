```python
# Number 12 

def compute(v):
    """
    Piecewise function:
    f(x) = 2*x       if x > 0
           1 - x     otherwise
    """
    v = np.array(v)
    return np.where(v > 0, 2*v, 1-v)

```

```python
v = np.array([-2, -1, 0, 1, 2, 3])
result = compute(v)
print(f' f(v) = {result}')
```
